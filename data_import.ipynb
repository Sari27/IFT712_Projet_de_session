```python
import pandas as pd
```

```python
df_train = pd.read_csv('leaf-classification/train.csv/train.csv').set_index('id')
df_test = pd.read_csv('leaf-classification/test.csv/test.csv').set_index('id')
```

```python
# Affichage du dataframe d'entraînement
df_train
```

```python
# Affichage du dataframe de test
df_test
```

```python
# Affichage des différentes classes possibles
df_train['species'].unique()
```

```python
# Normalisation du dataframe d'entraînement
for column in df_train.drop(columns=['species']):
    df_train[column] = (df_train[column] - min(df_train[column])) / (max(df_train[column]) - min(df_train[column]))
    
# Affichage du dataframe d'entraînement normalisé
df_train
```

```python
# Normalisation du dataframe de tests
for column in df_test:
    df_test[column] = (df_test[column] - min(df_test[column])) / (max(df_test[column]) - min(df_test[column]))
    
# Affichage du dataframe de test normalisé
df_test
```
