```python
%run ../methodes/perceptron_definition.ipynb import Perceptron_
import pandas as pd

df_train = pd.read_csv('../leaf-classification/train.csv/train.csv').set_index('id')
df_test = pd.read_csv('../leaf-classification/test.csv/test.csv').set_index('id')
```

```python
# Le terme de régularisation (penalty) est l'une des 3 fonctions suivantes
penaltys = ['l1', 'l2', 'elasticnet']

# Le taux d'apprentissage (learning rate) doit être entre 0 et 1
learning_rates = np.linspace(0.0001, 1, 10)

# La constante multiplicateur (alpha) du terme de régularisation doit être > 0
alphas = np.linspace(0.0000001, 0.001, 10)
```

```python
# Initialisation des meilleurs paramètres
err_min = 100
best_penalty = 'Nothing'
best_learning_rate = -1
best_alpha = -2
```

```python
# Initialisation des tableaux pour l'analyse des données
df_l1_train = pd.DataFrame(0, index=alphas, columns=learning_rates)
df_l2_train = pd.DataFrame(0, index=alphas, columns=learning_rates)
df_elasticnet_train = pd.DataFrame(0, index=alphas, columns=learning_rates)
df_l1_test = pd.DataFrame(0, index=alphas, columns=learning_rates)
df_l2_test = pd.DataFrame(0, index=alphas, columns=learning_rates)
df_elasticnet_test = pd.DataFrame(0, index=alphas, columns=learning_rates)
```

```python
%%capture --no-display
from sklearn.model_selection import KFold

# Pour obtenir de bons résultats on utilise la k-fold cross-validation
kf = KFold(n_splits=10)

for penalty in penaltys:
    for learning_rate in learning_rates:
        for alpha in alphas:
            
            # Initialisation de l'erreur moyenne
            err_train = 0
            err_test = 0
            
            # k-fold cross-validation sur les données d'entraînement
            for train, test in kf.split(df_train):
                x_train = df_train.iloc[train].drop(columns=['species'])
                t_train = df_train.iloc[train]['species']
                x_test = df_train.iloc[test].drop(columns=['species'])
                t_test = df_train.iloc[test]['species']
            
                # Initialisation du perceptron
                perceptron = Perceptron_(penalty, learning_rate, alpha)

                # Entraînement
                perceptron.entrainement(x_train, t_train)

                # Calcul de l'erreur d'entraînement
                pred_train = perceptron.prediction(x_train)
                err_train += perceptron.erreur(t_train, pred_train)

                # Calcul de l'erreur de test
                pred_test = perceptron.prediction(x_test)
                err_test += perceptron.erreur(t_test, pred_test)

            # Calcul de l'erreur totale
            err_train /= 10
            err_test /= 10
            err = (err_train + err_test) / 2 

            # On garde les données en mémoire
            if penalty == 'l1':
                df_l1_train[learning_rate][alpha] = err_train
                df_l1_test[learning_rate][alpha] = err_test
            elif penalty == 'l2':
                df_l2_train[learning_rate][alpha] = err_train
                df_l2_test[learning_rate][alpha] = err_test
            elif penalty == 'elasticnet':
                df_elasticnet_train[learning_rate][alpha] = err_train
                df_elasticnet_test[learning_rate][alpha] = err_test
                
            # On met à jour les meilleurs paramètres
            if err < err_min:
                err_min = err
                best_penalty = penalty
                best_learning_rate = learning_rate
                best_alpha = alpha
```

```python
import matplotlib.pyplot as plt
# Représentation graphique

# Evolution de l'erreur moyenne en fonction de alpha
plt.subplot(2,3,1) # elasticnet
plt.plot(df_elasticnet_train.index, df_elasticnet_train.sum(axis=1), color='green', label='train')
plt.plot(df_elasticnet_test.index, df_elasticnet_test.sum(axis=1), color='red', label='test')
plt.xlabel('alpha')
plt.ylabel('erreur')
plt.legend(loc="upper left")
plt.title('elasticnet')

plt.subplot(2,3,2) # l1
plt.plot(df_l1_train.index, df_l1_train.sum(axis=1), color='green')
plt.plot(df_l1_test.index, df_l1_test.sum(axis=1), color='red')
plt.xlabel('alpha')
plt.ylabel('erreur')
plt.title('l1')

plt.subplot(2,3,3) # l2
plt.plot(df_l2_train.index, df_l2_train.sum(axis=1), color='green')
plt.plot(df_l2_test.index, df_l2_test.sum(axis=1), color='red')
plt.xlabel('alpha')
plt.ylabel('erreur')
plt.title('l2')

# Evolution de l'erreur moyenne en fonction du learning rate
plt.subplot(2,3,4) # elasticnet
plt.plot(df_elasticnet_train.columns, df_elasticnet_train.sum(axis=0), color='green')
plt.plot(df_elasticnet_test.columns, df_elasticnet_test.sum(axis=0), color='red')
plt.xlabel('learning rate')
plt.ylabel('erreur')
plt.title('elasticnet')

plt.subplot(2,3,5) # l1
plt.plot(df_l1_train.columns, df_l1_train.sum(axis=0), color='green')
plt.plot(df_l1_test.columns, df_l1_test.sum(axis=0), color='red')
plt.xlabel('learning rate')
plt.ylabel('erreur')
plt.title('l1')

plt.subplot(2,3,6) # l2
plt.plot(df_l2_train.columns, df_l2_train.sum(axis=0), color='green')
plt.plot(df_l2_test.columns, df_l2_test.sum(axis=0), color='red')
plt.xlabel('learning rate')
plt.ylabel('erreur')
plt.title('l2')
```
